4. Phân tích mối liên hệ giữa các biến

     4.1 Biểu đồ (scatter plot)

Biểu đồ phân tán (scatter plot) là một trong những công cụ hữu hiệu và phổ biến nhất trong phân tích dữ liệu, đặc biệt trong việc hiểu và minh họa mối quan hệ giữa hai biến số định lượng. Mục đích chính của biểu đồ phân tán là xác định xem liệu có một mối quan hệ tuyến tính, phi tuyến tính, hoặc không có mối quan hệ nào giữa hai biến. Khi sử dụng scatter plot, chúng ta có thể dễ dàng quan sát mối liên hệ và xác định các mô hình, xu hướng, và các điểm bất thường trong dữ liệu.

Biểu đồ phân tán (scatter plot) là một công cụ phân tích mạnh mẽ trong lĩnh vực chứng khoán, giúp nhà đầu tư hiểu rõ mối quan hệ giữa các biến số định lượng. Trong chứng khoán, các biến số này có thể bao gồm giá cổ phiếu, khối lượng giao dịch, tỷ suất sinh lợi, độ biến động, và nhiều yếu tố khác ảnh hưởng đến hiệu suất của cổ phiếu hoặc danh mục đầu tư. Scatter plot không chỉ giúp phát hiện các xu hướng trong dữ liệu mà còn xác định mối quan hệ tương quan giữa các biến và những điểm bất thường, từ đó đưa ra quyết định đầu tư tốt hơn.

Sử dụng scatter plot để phân tích mối quan hệ giữa các biến: giá mở cửa (open), giá cao nhất (high), giá thấp nhất (low), giá đóng cửa (close), và khối lượng giao dịch (volume).

Kết quả nhận được:

•	Mối quan hệ dương mạnh: Giữa giá mở cửa và giá đóng cửa, cũng như giữa giá cao nhất và giá thấp nhất.

•	Mối quan hệ không rõ ràng: Giữa giá mở cửa, giá đóng cửa với khối lượng giao dịch. Một số ngày có thể có khối lượng giao dịch cao nhưng giá không biến động nhiều, và ngược lại.

•	Phát hiện ngoại lệ (Outliers): Có thể tồn tại những phiên giao dịch mà giá mở cửa hoặc giá đóng cửa có sự chênh lệch lớn với các biến khác, có thể gây ra sự chú ý về các sự kiện bất thường hoặc tin tức.

Kết Luận:

Việc sử dụng biểu đồ phân tán để phân tích mối quan hệ giữa các biến giá cổ phiếu và khối lượng giao dịch sẽ giúp nhà đầu tư có cái nhìn trực quan về cách mà các yếu tố này tương tác với nhau. Kết quả phân tích sẽ hỗ trợ việc ra quyết định đầu tư, cho thấy sự liên kết giữa biến động giá và khối lượng giao dịch, từ đó tạo ra các chiến lược giao dịch hiệu quả hơn.

 4.2 Ma trận tương quan (correlation matrix)

Trong phân tích dữ liệu tài chính, ma trận tương quan (correlation matrix) đóng vai trò quan trọng trong việc hiểu mối quan hệ giữa các biến số. Đặc biệt, khi phân tích giá cổ phiếu, việc nắm bắt tương quan giữa các giá trị như giá mở cửa (open), giá cao nhất (high), giá thấp nhất (low), giá đóng cửa (close), và khối lượng giao dịch (volume) có thể cung cấp thông tin quan trọng về xu hướng thị trường và sự biến động của giá cổ phiếu.

Ma trận tương quan (correlation matrix) là một công cụ quan trọng trong phân tích dữ liệu, đặc biệt hữu ích trong việc khám phá mối quan hệ giữa các biến. Trong trường hợp này, chúng ta tập trung vào các biến giá chứng khoán bao gồm: giá mở cửa (open), giá cao nhất (high), giá thấp nhất (low), giá đóng cửa (close) và khối lượng giao dịch (volume). Các biến này đại diện cho các yếu tố cơ bản trong việc phân tích kỹ thuật và dự đoán giá chứng khoán trong tương lai.

4.3 Hệ số tương quan (correlation coefficient)

Là một chỉ số thống kê dùng để đo lường mức độ và hướng mối quan hệ giữa hai biến số. Giá trị của hệ số này nằm trong khoảng từ -1 đến 1:

•	Hệ số tương quan bằng 1 cho thấy hai biến có mối tương quan thuận hoàn hảo. Khi một biến tăng thì biến kia cũng tăng.

•	Hệ số tương quan bằng -1 cho thấy hai biến có mối tương quan nghịch hoàn hảo. Khi một biến tăng thì biến kia giảm.
•	Hệ số tương quan bằng 0 cho thấy không có mối tương quan tuyến tính giữa hai biến

Độ đồng biến (covariance)
Là một chỉ số thống kê dùng để đo lường mối quan hệ giữa hai biến số. Nó cho biết mức độ mà hai biến số biến động cùng nhau:
-  Đồng biến dương:  cho thấy khi giá cổ phiếu của một ngân hàng tăng, giá cổ phiếu của ngân hàng còn lại cũng có xu hướng tăng.
-  Đồng biến âm:  giá cổ phiếu của một ngân hàng tăng thì giá cổ phiếu của ngân hàng kia giảm.
-  Đồng biến bằng 0: không có mối liên hệ rõ ràng giữa các cổ phiếu.



6. Một số chủ đề nâng cao:

Lựa chọn một trong các chủ để nâng cao sau để phân tích dữ liệu: Phân tích thành phần chính (PCA), phân tích nhóm (cluster analysis), khai phá dữ liệu (data mining).

Nhóm chọn phân tích nhóm (cluster analysis) để phân tích giá cổ phiếu của ngân hàng Agribank, Bidv và Vietinbank.
Phân tích nhóm (Cluster Analysis) trong ngữ cảnh phân tích giá cổ phiếu là một kỹ thuật thống kê dùng để phân loại cổ phiếu thành các nhóm (clusters) dựa trên các đặc điểm hoặc chỉ số tài chính tương tự, như giá trị cổ phiếu, khối lượng giao dịch, biến động giá, hoặc các chỉ số tài chính khác. Mục tiêu của phương pháp này là giúp nhà đầu tư và phân tích thị trường phát hiện ra các mẫu và xu hướng, cũng như cải thiện các quyết định đầu tư.

Các Thuật Toán Phân Tích Nhóm Thường Dùng:
1.	K-Means:
o	Là thuật toán phổ biến nhất, phân chia dữ liệu thành k cụm dựa trên khoảng cách tới tâm cụm.
2.	Hierarchical Clustering (Phân cụm phân cấp):
o	Tạo ra một cây phân cấp cho thấy mối quan hệ giữa các cụm. Có thể chọn mức độ phân nhóm bằng cách cắt cây.

o	Tập trung vào mật độ các điểm để tìm ra các nhóm, có khả năng phát hiện các nhóm có hình dạng không đồng nhất và xử lý tốt các điểm nhiễu.
4.	Gaussian Mixture Models (GMM):
o	Mô hình hóa dữ liệu bằng cách sử dụng các thành phần Gaussian, cho phép các cụm có hình dạng khác nhau.

Kết Quả Nhận Được từ Phân Tích Nhóm trong Phân Tích Giá Cổ Phiếu
Sau khi thực hiện phân tích nhóm (Cluster Analysis) sử dụng các biến như giá mở cửa, giá cao nhất, giá thấp nhất, giá đóng cửa, và khối lượng giao dịch, bạn có thể rút ra một số kết quả quan trọng như sau:
1.	Phân Nhóm Cổ Phiếu:
Các cổ phiếu được phân nhóm thành những nhóm tương tự dựa trên hành vi giá và khối lượng giao dịch. Mỗi nhóm có thể đại diện cho các cổ phiếu có tính chất và xu hướng tương tự, chẳng hạn như nhóm cổ phiếu có giá tăng mạnh, nhóm cổ phiếu có biến động thấp, hoặc nhóm cổ phiếu có khối lượng giao dịch cao.
2.	Hiểu Biến Động Giá:
Kết quả phân tích cho phép nhận diện những cổ phiếu có sự biến động giá tương tự. Ví dụ, nếu nhóm cổ phiếu của của CTG, BID và AGB trong một khoảng thời gian cụ thể nằm trong cùng một khoảng giá và khối lượng giao dịch tương tự, điều này cho thấy chúng có thể chịu ảnh hưởng từ các yếu tố bên ngoài giống nhau, chẳng hạn như tin tức thị trường hoặc chính sách kinh tế.
3.	Xác Định Xu Hướng: Kết quả có thể giúp xác định các xu hướng tiềm ẩn trong thị trường.
4.	Ra Quyết Định Đầu Tư: Việc phân nhóm giúp tìm ra các cổ phiếu có tính chất tương tự, từ đó giúp đưa ra quyết định đầu tư thông minh hơn.
5.	Phát Hiện Các Cổ Phiếu Bất Thường: Kết quả cũng có thể chỉ ra những cổ phiếu có hành vi bất thường so với phần còn lại của nhóm. Điều này có thể giúp phát hiện những cơ hội đầu tư hoặc cảnh báo về các rủi ro tiềm ẩn.
6.	So Sánh Giữa Các Nhóm:
So sánh hiệu suất giữa các nhóm cổ phiếu khác nhau. Ví dụ, Giả sử trong vòng một năm qua, giá cổ phiếu của VietinBank đã tăng 20%, trong khi BIDV chỉ tăng 10% và Agribank tăng 15%. Điều này cho thấy VietinBank có xu hướng tăng trưởng mạnh mẽ hơn so với hai ngân hàng còn lại.

4. Phân tích mối liên hệ giữa các biến

     4.1 Biểu đồ (scatter plot)

Biểu đồ phân tán (scatter plot) là một trong những công cụ hữu hiệu và phổ biến nhất trong phân tích dữ liệu, đặc biệt trong việc hiểu và minh họa mối quan hệ giữa hai biến số định lượng. Mục đích chính của biểu đồ phân tán là xác định xem liệu có một mối quan hệ tuyến tính, phi tuyến tính, hoặc không có mối quan hệ nào giữa hai biến. Khi sử dụng scatter plot, chúng ta có thể dễ dàng quan sát mối liên hệ và xác định các mô hình, xu hướng, và các điểm bất thường trong dữ liệu.

Biểu đồ phân tán (scatter plot) là một công cụ phân tích mạnh mẽ trong lĩnh vực chứng khoán, giúp nhà đầu tư hiểu rõ mối quan hệ giữa các biến số định lượng. Trong chứng khoán, các biến số này có thể bao gồm giá cổ phiếu, khối lượng giao dịch, tỷ suất sinh lợi, độ biến động, và nhiều yếu tố khác ảnh hưởng đến hiệu suất của cổ phiếu hoặc danh mục đầu tư. Scatter plot không chỉ giúp phát hiện các xu hướng trong dữ liệu mà còn xác định mối quan hệ tương quan giữa các biến và những điểm bất thường, từ đó đưa ra quyết định đầu tư tốt hơn.

Sử dụng scatter plot để phân tích mối quan hệ giữa các biến: giá mở cửa (open), giá cao nhất (high), giá thấp nhất (low), giá đóng cửa (close), và khối lượng giao dịch (volume).

Kết quả nhận được:

•	Mối quan hệ dương mạnh: Giữa giá mở cửa và giá đóng cửa, cũng như giữa giá cao nhất và giá thấp nhất.

•	Mối quan hệ không rõ ràng: Giữa giá mở cửa, giá đóng cửa với khối lượng giao dịch. Một số ngày có thể có khối lượng giao dịch cao nhưng giá không biến động nhiều, và ngược lại.

•	Phát hiện ngoại lệ (Outliers): Có thể tồn tại những phiên giao dịch mà giá mở cửa hoặc giá đóng cửa có sự chênh lệch lớn với các biến khác, có thể gây ra sự chú ý về các sự kiện bất thường hoặc tin tức.

Kết Luận:

Việc sử dụng biểu đồ phân tán để phân tích mối quan hệ giữa các biến giá cổ phiếu và khối lượng giao dịch sẽ giúp nhà đầu tư có cái nhìn trực quan về cách mà các yếu tố này tương tác với nhau. Kết quả phân tích sẽ hỗ trợ việc ra quyết định đầu tư, cho thấy sự liên kết giữa biến động giá và khối lượng giao dịch, từ đó tạo ra các chiến lược giao dịch hiệu quả hơn.

 4.2 Ma trận tương quan (correlation matrix)

Trong phân tích dữ liệu tài chính, ma trận tương quan (correlation matrix) đóng vai trò quan trọng trong việc hiểu mối quan hệ giữa các biến số. Đặc biệt, khi phân tích giá cổ phiếu, việc nắm bắt tương quan giữa các giá trị như giá mở cửa (open), giá cao nhất (high), giá thấp nhất (low), giá đóng cửa (close), và khối lượng giao dịch (volume) có thể cung cấp thông tin quan trọng về xu hướng thị trường và sự biến động của giá cổ phiếu.

Ma trận tương quan (correlation matrix) là một công cụ quan trọng trong phân tích dữ liệu, đặc biệt hữu ích trong việc khám phá mối quan hệ giữa các biến. Trong trường hợp này, chúng ta tập trung vào các biến giá chứng khoán bao gồm: giá mở cửa (open), giá cao nhất (high), giá thấp nhất (low), giá đóng cửa (close) và khối lượng giao dịch (volume). Các biến này đại diện cho các yếu tố cơ bản trong việc phân tích kỹ thuật và dự đoán giá chứng khoán trong tương lai.

4.3 Hệ số tương quan (correlation coefficient)

Là một chỉ số thống kê dùng để đo lường mức độ và hướng mối quan hệ giữa hai biến số. Giá trị của hệ số này nằm trong khoảng từ -1 đến 1:

•	Hệ số tương quan bằng 1 cho thấy hai biến có mối tương quan thuận hoàn hảo. Khi một biến tăng thì biến kia cũng tăng.

•	Hệ số tương quan bằng -1 cho thấy hai biến có mối tương quan nghịch hoàn hảo. Khi một biến tăng thì biến kia giảm.
•	Hệ số tương quan bằng 0 cho thấy không có mối tương quan tuyến tính giữa hai biến

Độ đồng biến (covariance)
Là một chỉ số thống kê dùng để đo lường mối quan hệ giữa hai biến số. Nó cho biết mức độ mà hai biến số biến động cùng nhau:
-  Đồng biến dương:  cho thấy khi giá cổ phiếu của một ngân hàng tăng, giá cổ phiếu của ngân hàng còn lại cũng có xu hướng tăng.
-  Đồng biến âm:  giá cổ phiếu của một ngân hàng tăng thì giá cổ phiếu của ngân hàng kia giảm.
-  Đồng biến bằng 0: không có mối liên hệ rõ ràng giữa các cổ phiếu.



6. Một số chủ đề nâng cao:

Lựa chọn một trong các chủ để nâng cao sau để phân tích dữ liệu: Phân tích thành phần chính (PCA), phân tích nhóm (cluster analysis), khai phá dữ liệu (data mining).

Nhóm chọn phân tích nhóm (cluster analysis) để phân tích giá cổ phiếu của ngân hàng Agribank, Bidv và Vietinbank.
Phân tích nhóm (Cluster Analysis) trong ngữ cảnh phân tích giá cổ phiếu là một kỹ thuật thống kê dùng để phân loại cổ phiếu thành các nhóm (clusters) dựa trên các đặc điểm hoặc chỉ số tài chính tương tự, như giá trị cổ phiếu, khối lượng giao dịch, biến động giá, hoặc các chỉ số tài chính khác. Mục tiêu của phương pháp này là giúp nhà đầu tư và phân tích thị trường phát hiện ra các mẫu và xu hướng, cũng như cải thiện các quyết định đầu tư.

Các Thuật Toán Phân Tích Nhóm Thường Dùng:
1.	K-Means:
o	Là thuật toán phổ biến nhất, phân chia dữ liệu thành k cụm dựa trên khoảng cách tới tâm cụm.
2.	Hierarchical Clustering (Phân cụm phân cấp):
o	Tạo ra một cây phân cấp cho thấy mối quan hệ giữa các cụm. Có thể chọn mức độ phân nhóm bằng cách cắt cây.
3.	DBSCAN (Density-Based Spatial Clustering of Applications with Noise):
o	Tập trung vào mật độ các điểm để tìm ra các nhóm, có khả năng phát hiện các nhóm có hình dạng không đồng nhất và xử lý tốt các điểm nhiễu.
4.	Gaussian Mixture Models (GMM):
o	Mô hình hóa dữ liệu bằng cách sử dụng các thành phần Gaussian, cho phép các cụm có hình dạng khác nhau.

	Quy trình phân tích nhóm:
-	 Thu Thập Dữ Liệu
Thu thập dữ liệu giá cổ phiếu và các thông tin liên quan từ nguồn dữ liệu đã được cung cấp từ vietstock về ngân hàng Agribank, Vietinbank và Bidv
o	Giá mở cửa (open)
o	Giá cao nhất (high)
o	Giá thấp nhất (low)
o	Giá đóng cửa (close)
o	Khối lượng giao dịch (volume)
-	 Tiền Xử Lý Dữ Liệu: Làm sạch và chuẩn bị dữ liệu để phân tích.
-	 Lựa Chọn Thuật Toán Phân Tích Nhóm: Quyết định thuật toán phù hợp để phân tích nhóm.
o	K-Means: Nếu muốn phân nhóm theo số lượng cụm đã xác định.
o	Hierarchical Clustering: Nếu muốn có cái nhìn trực quan về mối quan hệ giữa các cổ phiếu.
o	DBSCAN: Nếu cần phân nhóm với hình dạng không đồng nhất và xử lý nhiễu.
o	Gaussian Mixture Models (GMM): Nếu muốn các cụm có hình dạng khác nhau.
- Thực Hiện Phân Tích Nhóm
Phân nhóm cổ phiếu dựa trên các chỉ số đã chọn, sử dụng thuật toán đã chọn để phân nhóm.
- Phân Tích Kết Quả
Phân tích kết quả sau khi chạy thuật toán để từ đó hiểu rõ các nhóm cổ phiếu đã phân, đánh giá các nhóm được tạo ra, tìm hiểu các đặc điểm chung của từng nhóm.
o	Ví dụ: Xem xét giá trị trung bình của các chỉ số cho mỗi nhóm để hiểu rõ hơn về hành vi cổ phiếu.
- Trực Quan Hóa Kết Quả
Trực quan hóa kết quả phân tích để dễ dàng hiểu và chia sẻ.
- Rút Ra Kết Luận và Đề Xuất: Tóm tắt các phát hiện và đề xuất hành động dựa trên phân tích.
•	Ví dụ:

o	Nhóm cổ phiếu có sự biến động tương tự có thể phản ứng giống nhau đối với các yếu tố bên ngoài.
o	Các cổ phiếu có tiềm năng tăng trưởng hoặc giảm rủi ro có thể được xác định dựa trên phân tích nhóm.
Kết Quả Nhận Được từ Phân Tích Nhóm trong Phân Tích Giá Cổ Phiếu
Sau khi thực hiện phân tích nhóm (Cluster Analysis) sử dụng các biến như giá mở cửa, giá cao nhất, giá thấp nhất, giá đóng cửa, và khối lượng giao dịch, bạn có thể rút ra một số kết quả quan trọng như sau:
1.	Phân Nhóm Cổ Phiếu:
Các cổ phiếu được phân nhóm thành những nhóm tương tự dựa trên hành vi giá và khối lượng giao dịch. Mỗi nhóm có thể đại diện cho các cổ phiếu có tính chất và xu hướng tương tự, chẳng hạn như nhóm cổ phiếu có giá tăng mạnh, nhóm cổ phiếu có biến động thấp, hoặc nhóm cổ phiếu có khối lượng giao dịch cao.
2.	Hiểu Biến Động Giá:
Kết quả phân tích cho phép nhận diện những cổ phiếu có sự biến động giá tương tự. Ví dụ, nếu nhóm cổ phiếu của của CTG, BID và AGB trong một khoảng thời gian cụ thể nằm trong cùng một khoảng giá và khối lượng giao dịch tương tự, điều này cho thấy chúng có thể chịu ảnh hưởng từ các yếu tố bên ngoài giống nhau, chẳng hạn như tin tức thị trường hoặc chính sách kinh tế.
3.	Xác Định Xu Hướng: Kết quả có thể giúp xác định các xu hướng tiềm ẩn trong thị trường.
4.	Ra Quyết Định Đầu Tư: Việc phân nhóm giúp tìm ra các cổ phiếu có tính chất tương tự, từ đó giúp đưa ra quyết định đầu tư thông minh hơn.
5.	Phát Hiện Các Cổ Phiếu Bất Thường: Kết quả cũng có thể chỉ ra những cổ phiếu có hành vi bất thường so với phần còn lại của nhóm. Điều này có thể giúp phát hiện những cơ hội đầu tư hoặc cảnh báo về các rủi ro tiềm ẩn.
6.	So Sánh Giữa Các Nhóm:
So sánh hiệu suất giữa các nhóm cổ phiếu khác nhau. Ví dụ, Giả sử trong vòng một năm qua, giá cổ phiếu của VietinBank đã tăng 20%, trong khi BIDV chỉ tăng 10% và Agribank tăng 15%. Điều này cho thấy VietinBank có xu hướng tăng trưởng mạnh mẽ hơn so với hai ngân hàng còn lại.

library(quantmod)

# Định nghĩa thời gian
start_date <- as.Date("2024-01-02")
end_date <- as.Date("2024-06-28")

# Tải dữ liệu cho từng mã cổ phiếu
symbols <- c("AGR.VN", "CTG.VN", "BID.VN")

data_list <- lapply(symbols, function(sym) {
  getSymbols(sym, src = "yahoo", from = start_date, to = end_date, auto.assign = FALSE)
})

# Đặt tên cho danh sách dữ liệu

#chuyển đổi dữ liệu thành dataframe

write.csv(agr_df, "AGR_data.csv", row.names = FALSE)
write.csv(ctg_df, "CTG_data.csv", row.names = FALSE)

**lapply(symbols, function(sym) {...}):** Sử dụng lapply để lặp qua từng mã cổ phiếu trong vector symbols. Đối với mỗi mã cổ phiếu, nó thực hiện một hàm để lấy dữ liệu.

**getSymbols(sym, src = "yahoo", from = start_date, to = end_date, auto.assign = FALSE):**

**sym:** Mã cổ phiếu hiện tại trong vòng lặp.

**src = "yahoo":** Dữ liệu sẽ được lấy từ Yahoo Finance.

**from = start_date:** Ngày bắt đầu lấy dữ liệu.

**to = end_date:** Ngày kết thúc lấy dữ liệu.

**auto.assign = FALSE:** Không tự động gán dữ liệu vào biến có tên tương ứng với mã cổ phiếu. Thay vào đó, dữ liệu được lưu vào danh sách data_list.

````python
# Hàm làm sạch dữ liệu
clean_data <- function(df) {
  # Loại bỏ các giá trị NA
  df <- na.omit(df)

  # Đổi tên cột
  colnames(df) <- c("Open", "High", "Low", "Close", "Volume", "Adjusted_Close", "Date")

  # Chọn cột cần thiết
  df <- df[, c("Date", "Open", "High", "Low", "Close", "Volume", "Adjusted_Close")]

  return(df)
}

# Làm sạch dữ liệu cho từng mã cổ phiếu
agr_clean <- clean_data(agr_df)
ctg_clean <- clean_data(ctg_df)
bid_clean <- clean_data(bid_df)

# Xem dữ liệu đã làm sạch
head(agr_clean)
head(ctg_clean)
head(bid_clean)

# lưu dữ liệu đã làm sạch
write.csv(agr_clean, "AGR_clean_data.csv", row.names = FALSE)
write.csv(ctg_clean, "CTG_clean_data.csv", row.names = FALSE)
write.csv(bid_clean, "BID_clean_data.csv", row.names = FALSE)
````

Ứng dụng: Nhận xét từ biểu đồ có thể giúp đưa ra các quyết định trong phân tích giá trị nông nghiệp hoặc dự đoán xu hướng trong tương lai.

````python
# Tính CDF (Cumulative Distribution Function) cho mỗi nhóm "Chi_sô"
# Sắp xếp giá trị "Giá_tri" theo thứ tự tăng dần và tính tỷ lệ tích lũy
agri_long_sorted <- agri_long %>%
  dplyr::arrange(Giá_tri) %>%
  dplyr::group_by(Chi_sô) %>%
  dplyr::mutate(CDF = cumsum(rep(1, length(Giá_tri))) / length(Giá_tri))  # Tính CDF

# Vẽ biểu đồ CDF với ggplot2
ggplot(agri_long_sorted, aes(x = Giá_tri, y = CDF, color = Chi_sô, group = Chi_sô)) +
  geom_step(size = 1.5) +  # Vẽ biểu đồ bước
  labs(title = "CDF Cua Các Giá Tri Thông Kê Ngân Hàng AGR",
       x = "Giá tri", y = "Xác suât Tích Lũy (CDF)") +
  scale_color_brewer(palette = "Set2") +  # Bảng màu Set2
  theme_minimal(base_size = 15) +  # Theme đơn giản
  theme(legend.position = "top")  # Chú giải ở trên
````

Biểu đồ CDF thể hiện sự phân bố tích lũy của các giá trị trong dữ liệu. Khi bạn vẽ CDF, bạn sẽ thấy các đặc điểm sau:

Hình dạng CDF: Biểu đồ CDF sẽ có hình dạng "bước" (step function), trong đó mỗi bước tương ứng với một giá trị của Giá_tri trong dữ liệu. Điều này thể hiện sự tích lũy của xác suất cho các giá trị dữ liệu.

Tính phân tán: CDF cho phép bạn dễ dàng thấy mức độ phân tán của dữ liệu. Nếu CDF có một "bước" rộng, điều này có thể chỉ ra rằng có một nhóm giá trị lớn ở khu vực đó (ví dụ, các giá trị gần với X.OPEN hoặc X.HIGH).

So sánh giữa các nhóm: Bạn có thể dễ dàng so sánh sự phân bố của các nhóm như X.OPEN, X.HIGH, X.LOW, và X.CLOSE. Ví dụ, nhóm X.HIGH có thể có một phân bố tích lũy tăng chậm hơn so với nhóm X.LOW vì dữ liệu trong nhóm X.HIGH có giá trị cao hơn và sự phân bố rộng hơn.

Thông tin từ CDF:

CDF cho thấy xác suất tích lũy cho mỗi giá trị trong tập dữ liệu. Các "bước" trên đồ thị cho biết xác suất thấp nhất đến mức đó. Ví dụ, nếu CDF đạt 1.0 tại giá trị nào đó, điều này có nghĩa là tất cả các giá trị trong dữ liệu đều nhỏ hơn hoặc bằng giá trị đó.
Đánh giá sự đồng nhất hoặc phân tán của dữ liệu:

CDF sẽ cho bạn biết liệu dữ liệu có đồng nhất hay có sự phân tán rõ ràng. Nếu CDF dốc đứng (tăng nhanh), điều này có thể cho thấy dữ liệu tập trung ở các giá trị thấp hoặc trung bình. Nếu CDF khá phẳng, điều này chỉ ra sự phân tán đồng đều hơn giữa các giá trị.
4. Kết Luận
Tính ổn định của giá trị: Các nhóm giá trị như X.CLOSE và X.OPEN có sự phân phối khá đồng đều và ổn định hơn so với các nhóm giá trị X.LOW và X.HIGH, trong đó X.HIGH có sự biến động lớn hơn.

Biến động trong các nhóm: CDF có thể giúp nhận thấy các nhóm có mức biến động cao hơn (như X.HIGH) và các nhóm ổn định hơn (như X.OPEN).

Tính tích lũy: Biểu đồ CDF là công cụ mạnh mẽ để xem xét sự tích lũy của các giá trị trong dữ liệu và có thể giúp phát hiện các khu vực có sự tập trung hoặc phân tán cao.

````python
# Vẽ CDF (Cumulative Distribution Function)
ggplot(bidv_long, aes(x = Giá_tri, color = Chi_sô)) +
  stat_ecdf(geom = "step") +  # Vẽ CDF bằng bước
  labs(title = "Xác Suât Tích Lũy (CDF) Các Giá tri Thông kê Ngân hàng BIDV",
       x = "Giá tri",
       y = "Xác suât Tích Lũy") +
  scale_color_brewer(palette = "Set1") +  # Chọn bảng màu
  theme_minimal() +
  theme(
    axis.title.x = element_text(size = 14),
    axis.title.y = element_text(size = 14),
    plot.title = element_text(size = 16, hjust = 0.5)
  )
````

Đồ thị CDF thể hiện sự phân bố của các giá trị trong mỗi chỉ số (Open, High, Low, Close).
Đường CDF cho thấy xác suất tích lũy của các giá trị không vượt quá một mức nhất định. Khi giá trị trên trục x tăng, đường CDF cũng dần tăng, cho thấy xác suất của các giá trị nhỏ hơn hoặc bằng giá trị đó.
Tính chất phân bố:
Open và Close có sự tích lũy tương đối đồng đều, thể hiện rằng các giá trị trong khoảng giá trị này phân bố ổn định.
High và Low có sự tăng xác suất nhanh hơn ở các giá trị thấp hơn, cho thấy rằng chúng có sự tập trung nhiều ở các mức giá thấp và có sự biến động mạnh hơn ở mức cao.
Sự khác biệt rõ rệt trong hình dạng của các đường CDF cho thấy High và Low có sự dao động lớn hơn so với Open và Close.
Tóm lại, đồ thị CDF giúp nhận diện được sự phân bố của các giá trị trong các chỉ số và xác định được mức giá nào có xác suất xuất hiện cao nhất.

````python
install.packages("corrplot")
library(corrplot)
# Chuyển đổi dữ liệu thành dataframe và thêm cột ngày tháng
agr_df <- as.data.frame(data_list$`AGR.VN`)
ctg_df <- as.data.frame(data_list$`CTG.VN`)
bid_df <- as.data.frame(data_list$`BID.VN`)

agr_df$Date <- index(data_list$`AGR.VN`)
ctg_df$Date <- index(data_list$`CTG.VN`)
bid_df$Date <- index(data_list$`BID.VN`)

# Đổi tên cột và loại bỏ NA
clean_data <- function(df) {
  colnames(df) <- c("Open", "High", "Low", "Close", "Volume")
  df <- na.omit(df) # Loại bỏ NA
  return(df)
}

# Làm sạch dữ liệu cho từng mã cổ phiếu
agr_clean <- clean_data(agr_df)
ctg_clean <- clean_data(ctg_df)
bid_clean <- clean_data(bid_df)

# Tính toán ma trận tương quan cho từng tập dữ liệu
cor_agr <- cor(agr_clean[, c("Open", "High", "Low", "Close", "Volume" )])
cor_ctg <- cor(ctg_clean[, c("Open", "High", "Low", "Close", "Volume" )])
cor_bid <- cor(bid_clean[, c("Open", "High", "Low", "Close", "Volume" )])

# Vẽ ma trận cho AGR
corrplot(cor_agr, method = "color",
         col = colorRampPalette(c("grey", "white", "#E6E6FA"))(200),
         tl.col = "black", tl.srt = 45,
         addCoef.col = "black")

# Thêm tiêu đề ở trên biểu đồ
title(main = "Ma trận tương quan của Agribank", line = 3, cex.main = 1.5, font.main = 2)

# Vẽ ma trận cho Vietinbank
corrplot(cor_ctg, method = "color",
         col = colorRampPalette(c("lightpink", "#BC8F8F", "#FFE4E1"))(200),
         tl.col = "black", tl.srt = 45,
         addCoef.col = "black")
# Thêm tiêu đề ở trên biểu đồ
title(main = "Ma trận tương quan của Vietinbank", line = 3, cex.main = 1.5, font.main = 2)

# Vẽ ma trận cho BIDV
corrplot(cor_bid, method = "color",
         col = colorRampPalette(c("#EEE8AA", "pink", "lightblue"))(200),
         tl.col = "black", tl.srt = 45,
         addCoef.col = "black")
# Thêm tiêu đề ở trên biểu đồ
title(main = "Ma trận tương quan của BIDV", line = 3, cex.main = 1.5, font.main = 2)


````

````python
import matplotlib.pyplot as plt

# Thêm cột Stock vào mỗi DataFrame
AGR_df['Stock'] = 'AGR'
BID_df['Stock'] = 'BID'
CTG_df['Stock'] = 'CTG'

# Kết hợp các DataFrame
combined_df = pd.concat([AGR_df, BID_df, CTG_df], ignore_index=True)

# Chuyển đổi cột Date thành datetime
combined_df['Date'] = pd.to_datetime(combined_df['Date'])

# Vẽ xu hướng cho giá cổ phiếu đóng cửa đã điều chỉnh
plt.figure(figsize=(12, 6))
for stock in combined_df['Stock'].unique():
    stock_data = combined_df[combined_df['Stock'] == stock]
    plt.plot(stock_data['Date'], stock_data['Adjusted_Close'], label=stock)

plt.title('Xu hướng của giá đóng cửa điều chỉnh')
plt.xlabel('Date')
plt.ylabel('Giá đóng cửa điều chỉnh')
plt.legend(title='Stock')
plt.grid(True)
plt.show()
````

* thực hiện phân tích cụm (cluster analysis) dựa trên giá đóng cửa đã điều chỉnh

````python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Tạo DataFrame mới df_cluster chỉ chứa cột Date và Adjusted_Close
df_cluster = combined_df[['Date', 'Adjusted_Close']].copy()

# Chuẩn hóa dữ liệu trong df_cluster
scaler = StandardScaler()
df_cluster['Adjusted_Close_Scaled'] = scaler.fit_transform(df_cluster[['Adjusted_Close']])

# Khởi tạo mô hình KMeans với số cụm bằng 3
kmeans = KMeans(n_clusters=3, random_state=0)

# Thực hiện phân tích cụm
df_cluster['Cluster'] = kmeans.fit_predict(df_cluster[['Adjusted_Close_Scaled']])

# In ra nhãn cụm
print("Nhãn cụm:")
print(df_cluster['Cluster'].unique())

# In ra trung tâm các cụm
cluster_centers = scaler.inverse_transform(kmeans.cluster_centers_)
print("\nTrung tâm các cụm:")
print(cluster_centers)
````

# B. Phân tích độ ổn định

````python

import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Đọc dữ liệu từ các file CSV
BID_df = pd.read_csv("/content/BID-20241007.txt")

AGR_df = pd.read_csv("/content/AGR-20241007.txt")
CTG_df = pd.read_csv("/content/CTG-20241007.txt")
# Hiển thị thông tin về DataFrame (tên cột và kiểu dữ liệu)
print("BID DataFrame Info:")
print(BID_df.info())
print("\nAGR DataFrame Info:")
print(AGR_df.info())
print("\nCTG DataFrame Info:")
print(CTG_df.info())
````

````python
# Tính toán Lợi nhuận (Return) từ giá đóng cửa
BID_df['Return'] = BID_df['<CLOSE>'].pct_change()  # Tính tỷ lệ thay đổi hàng ngày
AGR_df['Return'] = AGR_df['<CLOSE>'].pct_change()
CTG_df['Return'] = CTG_df['<CLOSE>'].pct_change()

# Tính toán Biến động (Volatility) – độ lệch chuẩn của lợi nhuận trong 30 ngày
BID_df['Volatility'] = BID_df['Return'].rolling(window=30).std()
AGR_df['Volatility'] = AGR_df['Return'].rolling(window=30).std()
CTG_df['Volatility'] = CTG_df['Return'].rolling(window=30).std()

# Lấy các cột cần thiết cho phân nhóm
features = ['<CLOSE>', 'Volatility', 'Return']

# Loại bỏ NaN từ các cột tính toán
BID_df = BID_df.dropna(subset=features)
AGR_df = AGR_df.dropna(subset=features)
CTG_df = CTG_df.dropna(subset=features)

# Chuẩn hóa dữ liệu
scaler = StandardScaler()

# Chuẩn hóa dữ liệu cho mỗi DataFrame
BID_scaled = scaler.fit_transform(BID_df[features])
AGR_scaled = scaler.fit_transform(AGR_df[features])
CTG_scaled = scaler.fit_transform(CTG_df[features])

# Áp dụng phân tích nhóm KMeans cho mỗi ngân hàng (sử dụng số nhóm n_clusters = 2)
kmeans_bid = KMeans(n_clusters=2, random_state=42)
kmeans_agr = KMeans(n_clusters=2, random_state=42)
kmeans_ctg = KMeans(n_clusters=2, random_state=42)

# Fit KMeans
BID_df['Cluster'] = kmeans_bid.fit_predict(BID_scaled)
AGR_df['Cluster'] = kmeans_agr.fit_predict(AGR_scaled)
CTG_df['Cluster'] = kmeans_ctg.fit_predict(CTG_scaled)

# Kiểm tra kết quả phân nhóm
print("\nBID DataFrame with Clusters:")
print(BID_df[['<TICKER>', 'Cluster']])

print("\nAGR DataFrame with Clusters:")
print(AGR_df[['<TICKER>', 'Cluster']])

print("\nCTG DataFrame with Clusters:")
print(CTG_df[['<TICKER>', 'Cluster']])
````

````python
# Vẽ biểu đồ phân nhóm
plt.figure(figsize=(15, 5))

# Biểu đồ scatter cho BID
plt.subplot(1, 3, 1)
plt.scatter(BID_scaled[:, 0], BID_scaled[:, 1], c=BID_df['Cluster'], cmap='viridis')
plt.title('Phân nhóm BID', fontsize=14)
plt.xlabel('Giá cổ phiếu (Price)', fontsize=12)
plt.ylabel('Biến động (Volatility)', fontsize=12)

# Biểu đồ scatter cho AGR
plt.subplot(1, 3, 2)
plt.scatter(AGR_scaled[:, 0], AGR_scaled[:, 1], c=AGR_df['Cluster'], cmap='viridis')
plt.title('Phân nhóm AGR', fontsize=14)
plt.xlabel('Giá cổ phiếu (Price)', fontsize=12)
plt.ylabel('Biến động (Volatility)', fontsize=12)

# Biểu đồ scatter cho CTG
plt.subplot(1, 3, 3)
plt.scatter(CTG_scaled[:, 0], CTG_scaled[:, 1], c=CTG_df['Cluster'], cmap='viridis')
plt.title('Phân nhóm CTG', fontsize=14)
plt.xlabel('Giá cổ phiếu (Price)', fontsize=12)
plt.ylabel('Biến động (Volatility)', fontsize=12)

# Hiển thị biểu đồ
plt.tight_layout()
plt.show()
````

# D. Đánh giá yếu tố ảnh hưởng đến biến động giá và kết luận

````python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Đọc dữ liệu
def load_data():
    file_paths = {
        "AGR": "E:/Data ĐA1/AGR_clean_data.csv",
        "BID": "E:/Data ĐA1/BID_clean_data.csv",
        "CTG": "E:/Data ĐA1/CTG_clean_data.csv"
    }
    dataframes = {key: pd.read_csv(path) for key, path in file_paths.items()}
    return dataframes

# Tiền xử lý dữ liệu
def preprocess_data(dataframes):
    for key in dataframes:
        dataframes[key]['Date'] = pd.to_datetime(dataframes[key]['Date'])

    combined_data = pd.concat([df.assign(Stock=key) for key, df in dataframes.items()],
                            ignore_index=True)

    # Chuẩn hóa dữ liệu
    numeric_columns = ['Open', 'High', 'Low', 'Close', 'Volume', 'Adjusted_Close']
    scaler = StandardScaler()
    combined_data_scaled = scaler.fit_transform(combined_data[numeric_columns])

    return combined_data, combined_data_scaled, numeric_columns

# phân cụm K-means
def perform_clustering(combined_data_scaled, k=3):
    kmeans = KMeans(n_clusters=k, random_state=42)
    clusters = kmeans.fit_predict(combined_data_scaled)
    return kmeans, clusters

# Phân tích và hiển thị kết quả
def analyze_results(combined_data, clusters, kmeans, numeric_columns):
    # Thêm nhãn cụm vào dữ liệu
    results_df = combined_data.copy()
    results_df['Cluster'] = clusters

    # Phân tích các đặc điểm của từng cụm
    cluster_stats = results_df.groupby(['Cluster', 'Stock']).agg({
        'Close': ['mean', 'std'],
        'Volume': ['mean', 'std']
    }).round(2)

    # Vẽ biểu đồ phân phối cụm
    plt.figure(figsize=(12, 6))
    sns.scatterplot(data=results_df, x='Close', y='Volume', hue='Cluster', style='Stock')
    plt.title('Phân cụm dựa trên Giá đóng cửa và Khối lượng giao dịch')
    plt.show()

    return cluster_stats

# Hàm main để chạy toàn bộ quá trình
def main():
    # Tải dữ liệu
    dataframes = load_data()

    # Tiền xử lý dữ liệu
    combined_data, combined_data_scaled, numeric_columns = preprocess_data(dataframes)

    # Thực hiện phân cụm
    kmeans, clusters = perform_clustering(combined_data_scaled)

    # Phân tích kết quả
    cluster_stats = analyze_results(combined_data, clusters, kmeans, numeric_columns)

    print("\nThống kê cho từng cụm:")
    print(cluster_stats)

if __name__ == "__main__":
    main()
````

**Phân tích cụm để đưa ra nhận định về các yếu tố ảnh hưởng.**

````python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Định nghĩa lớp phân tích cụm
class ClusterAnalyzer:
    def __init__(self, dataframes):
        self.dataframes = dataframes
        self.combined_data = None
        self.clusters = None
        self.kmeans = None

    def analyze_clusters(self):
        # Chuẩn bị dữ liệu
        self.prepare_data()

        # phân cụm K-Means
        self.perform_clustering(k=3)

        # phân tích chi tiết cho từng cụm
        self.cluster_characteristics()
        self.price_movement_analysis()
        self.volume_impact_analysis()
        self.volatility_analysis()

    def prepare_data(self):
        # Kết hợp dữ liệu và thêm các chỉ báo mới
        combined = []
        for stock, df in self.dataframes.items():
            temp_df = df.copy()
            # biến động giá
            temp_df['Price_Change'] = temp_df['Close'].pct_change()
            # biến động khối lượng
            temp_df['Volume_Change'] = temp_df['Volume'].pct_change()
            # Tính Volatility
            temp_df['Volatility'] = (temp_df['High'] - temp_df['Low']) / temp_df['Open']
            temp_df['Stock'] = stock
            combined.append(temp_df)

        self.combined_data = pd.concat(combined, ignore_index=True)
        self.combined_data.dropna(inplace=True)  # Loại bỏ các hàng có giá trị NaN

    def perform_clustering(self, k=3):
        # Tiền xử lý dữ liệu
        numeric_columns = ['Open', 'High', 'Low', 'Close', 'Volume', 'Adjusted_Close']
        scaler = StandardScaler()
        combined_data_scaled = scaler.fit_transform(self.combined_data[numeric_columns])

        # phân cụm K-Means
        self.kmeans = KMeans(n_clusters=k, random_state=42)
        self.clusters = self.kmeans.fit_predict(combined_data_scaled)

        # Thêm nhãn cụm vào dữ liệu
        self.combined_data['Cluster'] = self.clusters

    def cluster_characteristics(self):
        print("\n=== PHÂN TÍCH ĐẶC ĐIỂM CỦA TỪNG CỤM ===")

        # Thống kê cơ bản cho từng cụm
        cluster_stats = self.combined_data.groupby('Cluster').agg({
            'Price_Change': ['mean', 'std'],
            'Volume_Change': ['mean', 'std'],
            'Volatility': ['mean', 'std']
        }).round(4)

        print("\nThống kê đặc điểm các cụm:")
        print(cluster_stats)

        # Vẽ biểu đồ phân phối biến động giá theo cụm
        plt.figure(figsize=(12, 6))
        sns.boxplot(x='Cluster', y='Price_Change', data=self.combined_data)
        plt.title('Phân phối biến động giá theo cụm')
        plt.show()

    def price_movement_analysis(self):
        print("\n=== PHÂN TÍCH XU HƯỚNG GIÁ ===")

        # Tính tương quan giữa biến động giá và khối lượng cho từng cụm
        for cluster in self.combined_data['Cluster'].unique():
            cluster_data = self.combined_data[self.combined_data['Cluster'] == cluster]
            correlation = cluster_data['Price_Change'].corr(cluster_data['Volume_Change'])
            print(f"Cụm {cluster}: Hệ số tương quan giữa biến động giá và khối lượng: {correlation:.4f}")

    def volume_impact_analysis(self):
        print("\n=== PHÂN TÍCH ẢNH HƯỞNG CỦA KHỐI LƯỢNG GIAO DỊCH ===")

        # Phân tích khối lượng giao dịch bất thường
        for cluster in self.combined_data['Cluster'].unique():
            cluster_data = self.combined_data[self.combined_data['Cluster'] == cluster]
            abnormal_volume = np.sum((cluster_data['Volume'] - cluster_data['Volume'].mean()).abs() > 2 * cluster_data['Volume'].std())
            print(f"Cụm {cluster}: Số lượng phiên có khối lượng giao dịch bất thường: {abnormal_volume}")

    def volatility_analysis(self):
        print("\n=== PHÂN TÍCH ĐỘ BIẾN ĐỘNG ===")

        # Phân tích độ biến động theo cụm
        volatility_stats = self.combined_data.groupby('Cluster')['Volatility'].agg([
            'mean', 'std', 'min', 'max'
        ]).round(4)

        print("\nThống kê độ biến động theo cụm:")
        print(volatility_stats)

        # Vẽ biểu đồ độ biến động
        plt.figure(figsize=(12, 6))
        sns.boxplot(x='Cluster', y='Volatility', data=self.combined_data)
        plt.title('Độ biến động theo cụm')
        plt.show()

def main():
    # Đọc dữ liệu
    dataframes = {
        'AGR': pd.read_csv("E:/Data ĐA1/AGR_clean_data.csv"),
        'BID': pd.read_csv("E:/Data ĐA1/BID_clean_data.csv"),
        'CTG': pd.read_csv("E:/Data ĐA1/CTG_clean_data.csv")
    }

    # Khởi tạo và chạy phân tích
    analyzer = ClusterAnalyzer(dataframes)
    analyzer.analyze_clusters()
if __name__ == "__main__":
    main()
````
